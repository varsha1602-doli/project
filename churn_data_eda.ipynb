```python
# churn_data_eda.ipynb

# ✅ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ✅ Step 2: Load Customer Data
df = pd.read_csv("../data/customer_data.csv")  # Adjust path if needed

# ✅ Step 3: Preview Data
print("First 5 rows:")
print(df.head())

print("\nData Info:")
print(df.info())

# ✅ Step 4: Basic Statistics
print("\nSummary Statistics:")
print(df.describe())

# ✅ Step 5: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# ✅ Step 6: Visualize Feature Distributions
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Premium'], kde=True)
plt.title("Premium Distribution")
plt.show()

# ✅ Step 7: Visualize Churn by Categorical Features
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='PolicyType', hue='Churn', data=df)
plt.title("Churn by Policy Type")
plt.show()

# ✅ Step 8: Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# ✅ Step 9: Conclusion
print("\nEDA Completed Successfully. Analyze charts and insights for further modeling.")
```
